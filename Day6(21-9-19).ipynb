### Visualizing the Data
- An intersting way to display Dataframe
- WIth matplotlib -- math plot library

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.show()
```

```python
marks = [79,81,67,98]

divisions = ['A','B','C','D']
index = np.arange(len(divisions))
plt.bar(divisions,marks,color="#1412cd",width = 0.7)
plt.show()
```

```python
plt.barh(divisions,marks,color='blue')
plt.show()
```

```python
lang = ['python','java','c','c++']
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
